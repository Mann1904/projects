## Data Preprocessing

```python
# importing libraries
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
```

```python
# read data csv files
deliverydf = pd.read_csv('../data/deliveries_data.csv')
matchdf = pd.read_csv('../data/match_data.csv')
```

```python
matchdf.match_id = matchdf.match_id.astype(int)
matchdf.head()
```

```python
deliverydf.wicket = deliverydf.wicket.astype(int)
deliverydf.match_id = deliverydf.match_id.astype(int)
deliverydf.head()
```

```python
#sum of missing values in each column
matchdf.isnull().sum()
```

```python
#sum of missing values in each column
deliverydf.isnull().sum()
```

```python
# replace value
matchdf = matchdf.replace({'Rising Pune Supergiant':'Rising Pune Supergiants','Delhi Daredevils':'Delhi Capitals'})
deliverydf = deliverydf.replace({'Rising Pune Supergiant':'Rising Pune Supergiants','Delhi Daredevils':'Delhi Capitals'})
```

```python
#merge season value from the matches dataframe to deliveries
data = matchdf.merge(deliverydf, left_on = 'match_id', right_on = 'match_id', how = 'left')
```

```python
data.head()
```

```python
# team percentage of matches won out of total matches played

df = data
# team matches played
team_matches_played = df.groupby(['match_id','batting_team'])['total_runs'].sum().reset_index()
team_matches_played['count']=1
team_matches_played = team_matches_played.groupby(['batting_team'])['count'].sum().reset_index()

# team innings win
df = matchdf[(matchdf.result==1)|(matchdf.result==2)]
df['match_wins'] = 1
team_matches_win = df.groupby(['winner'])['match_wins'].sum().reset_index()
team_matches_win = team_matches_win.merge(team_matches_played, left_on='winner', right_on='batting_team',how='left')
team_matches_win.rename(columns={'winner':'team','count':'matches_played'},inplace=True)
team_matches_win.drop(['batting_team'],axis=1,inplace=True)

# team win percentage
team_win_percentage = team_matches_win
team_win_percentage['win%'] = (team_win_percentage['match_wins']/team_win_percentage['matches_played'])*100
team_win_percentage.sort_values(ascending=False,by='win%')
```

```python
# add new column won% of batting team and bowling team
team_win_percentage.rename(columns={'team':'batting_team','win%':'batting_team_won%'},inplace=True)
data = data.merge(team_win_percentage[['batting_team','batting_team_won%']],left_on='batting_team',right_on='batting_team',how='left')
team_win_percentage.rename(columns={'batting_team':'bowling_team','batting_team_won%':'bowling_team_won%'},inplace=True)
data = data.merge(team_win_percentage[['bowling_team','bowling_team_won%']],left_on='bowling_team',right_on='bowling_team',how='left')

data.tail()
```

```python
# calculate final score of team
df = data
final_score = df.groupby(['match_id','batting_team'])['total_runs'].sum().reset_index()
final_score.rename(columns={'total_runs':'final_score'},inplace=True)
inning1 = final_score[final_score.index%2==0]
inning1['inning']=1
inning2 = final_score[final_score.index%2!=0]
inning2['inning']=2
inningdf = pd.concat([inning1,inning2],axis=0)

# inning 1 & 2 data
new_data = data.merge(inningdf,left_on=['match_id','batting_team'],right_on=['match_id','batting_team'],how='left')
new_data.head()
```

```python
# rename the column match_id to id
new_data.rename(columns={'match_id':'id'},inplace=True)
new_data
```

```python
result=new_data
```

```python
result.bowling_team.unique()
```

```python
# Keeping only consistent teams
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
                    'Delhi Capitals', 'Sunrisers Hyderabad']

result = result[(result['batting_team'].isin(consistent_teams)) & (result['bowling_team'].isin(consistent_teams))]
```

```python
dataset = result

number_of_time_win_match = dict()

for team1 in consistent_teams:
  for team2 in consistent_teams:
    if team1!=team2:
      set1 = (team1, team2)
      set1 = str(set1)
      if set1 not in number_of_time_win_match:
        number_of_time_win_match[set1] = dict()
        number_of_time_win_match[set1]['team1']=0
        number_of_time_win_match[set1]['team2']=0

id = 0
inning = 0
for id1, inn ,batting_team, bowling_team, winner in zip(dataset.id, dataset.inning, dataset.batting_team, dataset.bowling_team, dataset.winner):
  if (id!=id1 and inning!=inn) or (id==id1 and inning!=inn):
    set1 = (batting_team, bowling_team)
    set1 = str(set1)
    if batting_team==winner:
      number_of_time_win_match[set1]['team1']+=1
    elif bowling_team==winner:
      number_of_time_win_match[set1]['team2']+=1
    id = id1
    inning = inn

import json
with open('../data/num_of_time_win_match.json', 'w') as fp:
    json.dump(number_of_time_win_match, fp)
```

```python
# add new column in dataset

dataset = result

number_of_time_batting_team_win = []
number_of_time_bowling_team_win = []

for batting_team, bowling_team in zip(dataset.batting_team, dataset.bowling_team):
  set1 = (batting_team, bowling_team)
  set1 = str(set1)
  number_of_time_batting_team_win.append(number_of_time_win_match[set1]['team1'])
  number_of_time_bowling_team_win.append(number_of_time_win_match[set1]['team2'])

dataset['number_of_time_batting_team_win'] = number_of_time_batting_team_win
dataset['number_of_time_bowling_team_win'] = number_of_time_bowling_team_win

result = dataset
```

```python
# create new column for current score
score = [result.total_runs[0]]
id = 1
inning = 1
for run,ids,innings in zip(result.total_runs[1:], result.id[1:], result.inning[1:]):
    if ids==id and innings==inning:
        score.append(run+score[-1])
    else:
        score.append(run)
        id=ids
        inning=innings
result["score"] = score

result.head()
```

```python
# convert player name to general format

player = [result.batsman[0],result.non_striker[0]]
id = 1
inning = 1
for id1, inning1, batsman, non_striker, index in zip(result.id, result.inning, result.batsman, result.non_striker, result.index):
    
    if (id==id1 and inning==inning1):
        
        if batsman in player:
            n = player.index(batsman)+1
            result.batsman[index]='player'+str(n)
        else:
            player.append(batsman)
            n=player.index(batsman)+1
            result.batsman[index]='player'+str(n)
            
        if non_striker in player:
            n = player.index(non_striker)+1
            result.non_striker[index]='player'+str(n)
        else:
            player.append(non_striker)
            n=player.index(non_striker)+1
            result.non_striker[index]='player'+str(n)
    else:
        id=id1
        inning=inning1
        player.clear()
        player.append(result.batsman[index])
        player.append(result.non_striker[index])
        
        n=player.index(batsman)+1
        result.batsman[index]='player'+str(n)
            
        n=player.index(non_striker)+1
        result.non_striker[index]='player'+str(n)
```

```python
# calculate number of wickets gone
 
# fill null value with 0
result.wicket_kind = result.wicket_kind.fillna(0)

wickets = []
id = 1
inning = 1
wicket = 0
for id1, inning1, wckt in zip(result.id, result.inning, result.wicket_kind):
  if (id==id1 and inning==inning1):
    if wckt==0 :
      wickets.append(wicket)
    else:
      wicket+=1
      wickets.append(wicket)
  else:
    id = id1
    inning = inning1
    wicket = 0
    wickets.append(wicket)

result['wickets'] = wickets
```
